# Data Cleaning

```python
df["date_added"]=pd.to_datetime(df["date_added"],errors="coerce")
```

```python
df["director"]=df["director"].fillna("Unknown")
df["cast"]=df["cast"].fillna("Unknown")
df["country"]=df["country"].fillna("Unknown")
df["date_added"]=df["date_added"].ffill()
df["rating"]=df["rating"].fillna("Unknown")
df["duration"]=df["duration"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df["type"]=df["type"].str.title()
df["rating"]=df["rating"].str.upper()
df["country"]=df["country"].str.title()
```

# Univariate and Bivariate Analysis

```python
rate_category=df["rating"].value_counts()
rate_category=rate_category.idxmax()
print("The most common rating category is:",rate_category)
```

```python
country_distribution = df["country"].value_counts()
top_5_countries = country_distribution.head(5)
print(top_5_countries)
```

```python
avg_release=df.groupby("type")["release_year"].mean()
print(avg_release)
```

# Trend and Relationship Exploration


```python
df["year_added"]=df["date_added"].dt.year
titles=df.groupby("year_added")["title"].count()
titles
```

The analysis of the above result shows that there was significant increase in content added particularly from 2016 to 2019, followed by a slight decline in 2020 and 2021.

```python
sns.boxplot(x="rating",y="release_year",data=df)
plt.xticks(rotation=45)
plt.show()
```

# 5. Visualizations


```python
sns.countplot(x="type",data=df)
plt.title("Count of Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
top_5_countries=df["country"].value_counts().head(5)
top_5_countries.plot(kind="bar")
plt.title("Top 5 Countries with Most Content Added")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.lineplot(x=titles.index, y=titles.values)
plt.title('Number of Titles Added Per Year')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
df["duration_num"]=df["duration"].str.split().str[0]
df["duration_num"]=pd.to_numeric(df["duration_num"],errors="coerce")

sns.scatterplot(x="release_year",y="duration_num",data=df)
plt.title("Release Year vs Duration")
plt.xlabel("Release Year")
plt.ylabel("Duration")
plt.show()
```

```python
plt.figure(figsize=(7,5))
corr=df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# Key Insights

* The United States contributes the highest number of titles in the dataset.  
* A significant increase in content additions is observed after 2015.
* Movies are more in number compared to TV Shows.
* Most of the content is rated TV-MA and TV-14, which shows that Netflix mainly focuses on mature and teenage audiences.
